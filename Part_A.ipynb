1. Using loops, implement functions to compute:
○ Mean
○ Median
○ Variance for a given list of numbers

```python
def mean(arr):
    total = 0
    for x in arr:
        total += x
    return total / len(arr)

def median(arr):
    arr = sorted(arr)
    n = len(arr)
    if n % 2 == 1:
        return arr[n // 2]
    else:
        return (arr[n//2 - 1] + arr[n//2]) / 2

def variance(arr):
    m = mean(arr)
    total = 0
    for x in arr:
        total += (x - m) ** 2
    return total / len(arr)

list1 = [47, 55, 45, 85, 74, 63, 95, 42, 23]
print(mean(list1))
print(median(list1))
print(variance(list1))
```

2. Rewrite the above computations using:
○ Higher-order functions (map, filter, reduce where applicable)

```python
from functools import reduce

def mean_hof(arr):
    return reduce(lambda x, y: x + y, arr) / len(arr)

def variance_hof(arr):
    m = mean_hof(arr)
    return reduce(lambda acc, x: acc + (x - m) ** 2, arr, 0) / len(arr)

list = [47, 55, 45, 85, 74, 63, 95, 42, 23]
print(mean_hof(list))
#print(median(list))
print(variance_hof(list))
```

4. Implement a function to compute:
○ Standard deviation
○ Verify your result using NumPy

```python
import numpy as np
import math

def std_dev(arr):
    var = variance(arr)
    return math.sqrt(var)

data = [1, 2, 3, 4, 5]

print(std_dev(data))
print(np.std(data))  
```

5. Given two groups of sample data, compute:
A = [12,23,3,34,45,56,5]
B = [ 12,1,3,1,1,2,3,4,5,3,4]
○ Mean and variance of both groups
○ Difference in means

```python
A = [12,23,3,34,45,56,5]
B = [12,1,3,1,1,2,3,4,5,3,4]

mean_A = mean(A)
mean_B = mean(B)

var_A = variance(A)
var_B = variance(B)

diff = mean_A - mean_B

print(mean_A, var_A)
print(mean_B, var_B)
print("Difference:", diff)
```
